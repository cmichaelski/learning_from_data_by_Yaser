## Legendre Polynomials
[Legendre polynomials](http://en.wikipedia.org/wiki/Legendre_polynomials)

### Generate Legendre Polynomials
We can legendre polynomials uses Bonnet’s recursion formula:
$nP_{n}(x) = (2n-1)xP_{n-1}(x) - (n-1)P_{n-2}(x), \mbox{ given } P_0(x) = 1, P_1(x) = x$

```python
## lengendre polynomials generating example
def generate_legendre_polynomials(max_degree):
    """
    Create a list of legendre polynomials from degree of zero to degree of max_degree.
    """
    x, n = symbols('x n')
    legendre_polynomial_list = [1, x]

    def nth_order_legendre(nth):
        """
        Create nth legendre polynomials iteratively 
        according to Bonnet’s recursion formula.
        """
        if nth <= 1:
            return legendre_polynomial_list[nth]
        
        for i in range(2, nth + 1):
            g = legendre_polynomial_list[i - 1]
            h = legendre_polynomial_list[i - 2]
            p = 1/n * ((2 * n - 1) * x * g -  (n - 1) * h)
            legendre_polynomial_list.append(expand(p.subs(n, i)))
        
    nth_order_legendre(max_degree)
    return legendre_polynomial_list

legendre_polynomial_table_10 = generate_legendre_polynomials(10)  # legendre polynomial table

print('Legendre Polynomials from degree of 0 to degree of 10\n' + 
      '\n'.join([l.__str__() for l in legendre_polynomial_table_10]))
```

### Orthogonailty of Legendre Polynomials

An important property of the Legendre polynomials is that they are orthogonal with respect to the $L^2$ inner product on the interval $−1 ≤ x ≤ 1$:
$$ \int_{-1}^1 P_m(x)P_n(x)\,dx = \frac{2}{2n+1}\delta_{mn} $$
where $\delta_{mn} = 1 \mbox{ if } m = n$ and 0 otherwise. 

```python
p1, p2, p3 = legendre_polynomial_table_10[1:4]
print('integral for p1p2:', integrate(p1 * p2, (symbols('x'), -1, 1)))
print('integral for p1p3:', integrate(p1 * p3, (symbols('x'), -1, 1)))
print('integral for p1^2:', integrate(p1 * p1, (symbols('x'), -1, 1)))
print('integral for p2^2:', integrate(p2 * p2, (symbols('x'), -1, 1)))
print('integral for p3^2:', integrate(p3 * p3, (symbols('x'), -1, 1)))
```

```python
from numpy.random import uniform, standard_normal
from numpy.linalg import pinv
from week6 import Dataset


def generate_legendre_coefficients(max_degree):    
    """
    Initialize the coefficients of the legendre polynomia randomly 
    accroding to the target function formula.
    """
    alphas = standard_normal(max_degree + 1)
    
    # nomalize term. use the orthogonailty of legendre polynomials
    sum = 0
    for k in range(max_degree + 1):
        sum += 1 / (2*k + 1)
    Z = np.sqrt(1 / sum)
    return alphas * Z


def evaluate_legendre(coefficients, x):
    """
    Evaluate the legendre polynomial value with given coefficients at x
    """
    c = coefficients
    max_degree = c.shape[0] - 1
    p0 = np.ones(x.shape[0])
    p1 = x

    if max_degree == 0:
        return p0 * c[0]
    if max_degree == 1:
        return p0 * c[0] + p1 * c[1]
    else:
        f = p0 * c[0] + p1 * c[1]
        for i in range(2, max_degree + 1):
            p2 = 1/i * ((2*i - 1) * x * p1 - (i - 1) * p0)
            f += p2 * c[i]
            p0 = p1
            p1 = p2
        return f
    

def legendre_polynomial_transform(x, max_degree):
    """
    Return a matrix of which ith column is the value of legendre polynomial of degree i.
    """
    X = [np.ones(x.shape[0]), x]
    for i in range(2, max_degree + 1):
        pi = 1/i * ((2*i - 1) * x * X[i - 1] - (i - 1) * X[i - 2])
        X.append(pi)
    return np.vstack(X).T


def analytical_difference(c1, c2):
    """
    Compute the analytical different between two legendre polynomials.
    """
    s1 = c1.size
    s2 = c2.size

    if s1 > s2:
        diff = c1.copy()
        diff[:s2] -= c2
    else:
        diff = c2.copy()
        diff[:s1] -= c1

    sum = 0
    for k, d in enumerate(diff):
        sum += d ** 2 / (2*k + 1)
    return np.sqrt(sum)
        

class LegendreDataset(Dataset):
    """
    Generate a dataset with selecting x uniformly on [-1, 1) 
    and y equals a combination of legendre polynomials using given coefficients puls noise.
    """   
    def __init__(self, N, sigma, legendre_coefficients):
        max_degree = legendre_coefficients.shape[0] - 1
        x = uniform(-1, 1, N)
        y = evaluate_legendre(legendre_coefficients, x) + sigma * standard_normal(N)      
        self.coefficients = legendre_coefficients
        super(LegendreDataset, self).__init__(x, y)
        
    def get_coefficients(self):
        return self.coefficients

    def plot(self, ax=None):
        """
        Plot data set and target function.
        """
        if ax is None:
            fig, ax = plt.subplots()
        
        # draw points
        ax.scatter(self.X, self.y, color='black')
        # draw target function
        x = np.linspace(-1, 1, 50)
        y = evaluate_legendre(self.coefficients, x)
        ax.plot(x, y, color='red', label='target')


class LinearRegression:
    """
    An linear regression object.
    """
    def get_weight(self):
        return self.w

    def fit(self, X, y):
        """
        Fit the dataset.
        """
        assert y.shape[0] == X.shape[0]

        self.w = pinv(X.T.dot(X)).dot(X.T).dot(y)
 
    def plot(self, color='green', label=None, ax=None):
        """
        Plot fitted curve.
        """
        assert self.w is not None
        if ax is None:
            fig, ax = plt.subplots()
        if label is None:
            label = 'h'

        # plot fitted curve
        x = np.linspace(-1, 1, 50)
        y = np.dot(legendre_polynomial_transform(x, self.w.shape[0] - 1), self.w)
        subindex = np.logical_and(y <= 8, y >= -8)  # sclar y axis
        ax.plot(x[subindex], y[subindex], color=color, label=label)


target = generate_legendre_coefficients(20)
ds = LegendreDataset(20, 1, target)
fig, ax = plt.subplots()
ds.plot(ax=ax)
fit2 = LinearRegression()
fit2.fit(legendre_polynomial_transform(ds.get_X(), 2), ds.get_y())
fit2.plot(color='green', label='h_2', ax=ax)
fit10 = LinearRegression()
fit10.fit(legendre_polynomial_transform(ds.get_X(), 10), ds.get_y())
fit10.plot(color='blue', label='h_10', ax=ax)
ax.legend(loc=0)
analytical_difference(fit2.get_weight(), target), analytical_difference(fit10.get_weight(), target)
```

## Run Experiment

```python
# default parameter setup
default = {
    'N': (40, 120, 2),  # min, max, step
    'sigma2': (0, 2, 0.05),
    'qf': (1, 101, 2),
    'average_over_times': 1000
}

def compute_overfit(N, qf, sigma, times=1000):
    """
    Compute overfit with given parametes.
    """
    overfit = 0
    for i in range(times):
        target = generate_legendre_coefficients(qf)
        ds = LegendreDataset(N, sigma, target)
        fit2 = LinearRegression()
        fit2.fit(legendre_polynomial_transform(ds.get_X(), 2), ds.get_y())
        fit10 = LinearRegression()
        fit10.fit(legendre_polynomial_transform(ds.get_X(), 10), ds.get_y())
        overfit += (analytical_difference(fit10.get_weight(), target)
                    - analytical_difference(fit2.get_weight(), target))
    
    overfit /= times
    if overfit > 0.2:
        overfit = 0.2
    if overfit < -0.2:
        overfit = -0.2
    
    return overfit


def run_experiment_with_fix_qf(parameters, qf=20):
    N_min, N_max, N_step = parameters['N']
    sigma2_min, sigma2_max, sigma2_step = parameters['sigma2']
    N_n = np.floor((N_max - N_min) / N_step) + 1
    sigma2_n = np.floor((sigma2_max - sigma2_min) / sigma2_step) + 1
    e = np.zeros((sigma2_n, N_n))

    for i, sigma2 in enumerate(np.arange(sigma2_min, sigma2_max + sigma2_step, sigma2_step)):
        for j, N in enumerate(np.arange(N_min, N_max + N_step, N_step, dtype='int')):
            e[i, j] = compute_overfit(N, qf, np.sqrt(sigma2), parameters['average_over_times'])
    return e


def run_experiment_with_fix_sigma2(parameters, sigma2=0.1):
    N_min, N_max, N_step = parameters['N']
    qf_min, qf_max, qf_step = parameters['qf']
    N_n = np.floor((N_max - N_min) / N_step) + 1
    qf_n = np.floor((qf_max - qf_min) / qf_step) + 1
    e = np.zeros((qf_n, N_n))

    for i, qf in enumerate(np.arange(qf_min, qf_max + qf_step, qf_step, dtype='int')):
        for j, N in enumerate(np.arange(N_min, N_max + N_step, N_step, dtype='int')):
            e[i, j] = compute_overfit(N, qf, np.sqrt(sigma2), parameters['average_over_times'])
    return e
```

```python
params = default
params['average_over_times'] = 5000
e_sigma = run_experiment_with_fix_qf(params)
e_qf = run_experiment_with_fix_sigma2(params)
```

## Plot Results

```python
def plot_overfit(e, parameters, yaxis='sigma', ax=None):
    """
    e: the error mesurement matrix of overfit
    yaxis: sigma or qf
    """
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    
    N_min, N_max, N_step = parameters['N']
    sigma2_min, sigma2_max, sigma2_step = parameters['sigma2']
    qf_min, qf_max, qf_step = parameters['qf']

    if ax is None:
        fig = plt.figure()
        ax = fig.add_axes([0.2, 0.2, 0.7, 0.7])
    im = ax.imshow(e, origin='lower', interpolation='nearest', aspect='auto')
    
    # set x axis label
    ax.set_xticks(np.linspace(0, (N_max - N_min) // N_step, 5))
    ax.set_xticklabels(['%d' % (x * N_step + N_min)  for x in ax.get_xticks()])
    ax.set_xlabel(r'Number of data points, $N$')
    # set y axis label
    if yaxis == 'sigma':
        ax.set_yticks(np.linspace(0, (sigma2_max - sigma2_min) // sigma2_step, 5))
        ax.set_yticklabels(['%0.1f' % (y * sigma2_step + sigma2_min)  for y in ax.get_yticks()])
        ax.set_ylabel(r'Noise level $\sigma^2$')
        ax.set_title(r'Impact of $\sigma^2$')
    else:
        ax.set_yticks(np.linspace(0, (qf_max - qf_min) // qf_step, 5))
        ax.set_yticklabels(['%0.1f' % (y * qf_step + qf_min)  for y in ax.get_yticks()])
        ax.set_ylabel(r'Target complexity $Q_f$')
        ax.set_title(r'Impact of $Q_f$')

    # put colorbar at right
    divide = make_axes_locatable(ax)
    cax = divide.append_axes("right", size="5%", pad=0.1)
    plt.colorbar(im, cax=cax, ticks=[-0.2, -0.1, 0, 0.1, 0.2])

fig = plt.figure(figsize=(17, 8))
ax1 = plt.subplot2grid((1, 2), (0, 0), aspect=0.5)
ax2 = plt.subplot2grid((1, 2), (0, 1))
plot_overfit(e_sigma, params, ax=ax1)
plot_overfit(e_qf, params, yaxis='qf', ax=ax2)
```

The reason I think is that with the Qf increasing, $H_2$ has more and more deteriministic noise and it gets more overfitting, but for $H_{10}$ nothing changes. 

From the bais and variance view, e_out = var + bias(deterministic noise) + stochastic noise. Two hypotheses have the same quantity of stochasitc noise and $H_2$ is biased. $H_{10}$ is affected by the noise more than $H_2$. I don't how to prove it. But I have a intuition that each $g(x)$ of $H_{10}$ has a bigger swing from $\bar{g}(x)$ than $g(x)$ of $H_2$ because of higher variance, and the e_out of $H_{10}$ changes much more than the e_out of H_2 when the noise makes the swing even bigger.
